```python
def is_prime(x):
    for i in range(2, int(x/2)):
        if (x % i) == 0:
            return False
    
    return True



is_prime(100)
```

```python
def fibonacci_iterative(x):
    fiboNumPre = 0
    fiboNumRec = 1
    for i in range(2,x):
        fiboNumNew = fiboNumPre + fiboNumRec 
        fiboNumPre = fiboNumRec
        fiboNumRec = fiboNumNew
    return fiboNumNew

fibonacci_iterative(4)
```

```python
def factorial_recursive(n):
    if n == 1:
        return 1
    else:
        return n * factorial_recursive(n-1)
factorial_recursive(10)
```

```python
"""Fibonacci using recursive"""
prev = 0
rec  = 1
def fibonacci_recursive(x):
    global prev , rec
    current = prev + rec
    prev = rec
    rec = current
    if x == 1 or x == 2:
        #print (current)
        return x-1
    else:
        #print(x-1 ,'>>',current)
        
        return fibonacci_recursive(x-1) 
#print (f"{prev} and {rec}")   
fibo = fibonacci_recursive(4)
print (f"{prev} and {rec} and {current}") 

print(f"{fibo}")
```

```python
import time
stored_result = {}
def sum_to_n(n):
    start_time = time.perf_counter()
    result = 0
    for i in reversed(range(n)):
        if i+1 in stored_result:
           
            print('Stopping sum at %s because we have previously computed it' % str(i + 1))
            result += stored_result[i+1]
            break
        else:
            result += i + 1
    stored_result[n]  = result 
    print('Clalculating for',n,"takes",time.perf_counter() - start_time,"Seconds")
    return result


sum_to_n(5)
sum_to_n(7)
sum_to_n(1)
sum_to_n(12)

print(stored_result)
    
```

```python
"""
Fibonacci series using dynamic programming
and  timining it too.
"""

stored_fibo = {0:1,1:1}
def fibo_dynamic(x):
    if  x in stored_fibo:
        return stored_fibo[x]
    else:
        stored_fibo[x] = fibo_dynamic(x-2) + fibo_dynamic(x-1)
        return stored_fibo[x]
 

fibo_dynamic(4)
print(stored_fibo)
```
